# Load the Data

```python
import pandas as pd

df = pd.read_excel('df_mortality_regions.xlsx')
df.head()
```

# Islands

## Number of Islands 

```python
df.Island.sum()
```

## Which region had more Islands?

```python
df['Regional indicator']
```

```python
df['Regional indicator'].value_counts()
```

```python
mask = df.Island == 1
```

```python
df[mask]['Regional indicator'].value_counts()
```

```python
df.groupby(by='Regional indicator').Island.sum().sort_values(ascending=False)
```

## Show all Columns for these Islands

```python
df.iloc[[3, 14, 16, 22, 38]]
```

## Mean Age of across All Islands?

```python
df['Median age'].mean()
```

# Female Heads of State

## Number of Countries with Female Heads of State 

```python
df['Female head of government'].sum()
```

## Which region had more Female Heads of State?

```python
df.groupby(by='Regional indicator')['Female head of government'].sum().sort_values(ascending=False)
```

## Show all Columns for these Countries

```python
df.iloc[[0, 1, 3, 5, 12, 67]]
```

## Mean Age of across All Countries?

```python
df['Median age'].mean()
```

## Show All Columns for the Country with more deaths due to Covid per 100,000 people

```python
df.loc[mask,:]
```
